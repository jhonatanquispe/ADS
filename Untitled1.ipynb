```python
import math

# ---------------------------
# Función: f(x) = tan(x^2) - 0.5
# ---------------------------
def f(x):
    return math.tan(x**2) - 0.5

# ---------------------------
# a) Tabla de valores para graficar
# ---------------------------
print("TABLA DE VALORES PARA GRAFICAR (0 <= x <= 3):")
x = 0.0
while x <= 3.0:
    try:
        y = f(x)
        print(f"x = {round(x, 2)}, y = {round(y, 4)}")
    except:
        print(f"x = {round(x, 2)}, y = indefinido (valor fuera del dominio)")
    x += 0.1
print("\n")

# ---------------------------
# c) Método de bisección
# ---------------------------
def biseccion(a, b, error):
    if f(a) * f(b) >= 0:
        print("No se puede aplicar el método de bisección: f(a) y f(b) no tienen signos opuestos.")
        return None

    iteracion = 1
    while (b - a) / 2 > error:
        c = (a + b) / 2
        try:
            fc = f(c)
            print(f"Iteración {iteracion}: a = {round(a, 6)}, b = {round(b, 6)}, c = {round(c, 6)}, f(c) = {round(fc, 6)}")
            if fc == 0:
                return c
            elif f(a) * fc < 0:
                b = c
            else:
                a = c
        except:
            print(f"Iteración {iteracion}: f(c) indefinido en c = {c}")
        iteracion += 1
    return (a + b) / 2

# ---------------------------
# d) Raíz aproximada en [1.5, 2] con error 0.0001
# ---------------------------
print("RAÍZ EN INTERVALO [1.5, 2] CON ERROR 0.0001:")
raiz1 = biseccion(1.5, 2, 0.0001)
print(f"Raíz aproximada en [1.5, 2]: {raiz1}")
print("\n")

# ---------------------------
# e) Raíz aproximada en [2.5, 2.7] con error 0.0001
# ---------------------------
print("RAÍZ EN INTERVALO [2.5, 2.7] CON ERROR 0.0001:")
raiz2 = biseccion(2.5, 2.7, 0.0001)
print(f"Raíz aproximada en [2.5, 2.7]: {raiz2}")
```

```python
import math

# ---------------------------
# Función: f(x) = 2e^x - 7x^3 - 2
# ---------------------------
def f(x):
    return 2 * math.exp(x) - 7 * x**3 - 2

# ---------------------------
# a) Tabla de valores para graficar
# ---------------------------
print("TABLA DE VALORES PARA GRAFICAR (−1.5 <= x <= 1.5):")
x = -1.5
while x <= 1.5:
    try:
        y = f(x)
        print(f"x = {round(x, 2)}, y = {round(y, 4)}")
    except:
        print(f"x = {round(x, 2)}, y = indefinido")
    x += 0.1
print("\n")

# ---------------------------
# c) Método de Regla Falsa
# ---------------------------
def regla_falsa(a, b, error):
    if f(a) * f(b) >= 0:
        print("No se puede aplicar el método: f(a) y f(b) no tienen signos opuestos.")
        return None

    iteracion = 1
    c_anterior = a
    while True:
        fa = f(a)
        fb = f(b)
        c = (a * fb - b * fa) / (fb - fa)
        fc = f(c)

        print(f"Iteración {iteracion}: a = {round(a, 6)}, b = {round(b, 6)}, c = {round(c, 6)}, f(c) = {round(fc, 6)}")

        if abs(fc) < error or abs(c - c_anterior) < error:
            break

        if fa * fc < 0:
            b = c
        else:
            a = c

        c_anterior = c
        iteracion += 1

    return c

# ---------------------------
# d) Raíz en intervalo [-0.7, -0.3]
# ---------------------------
print("RAÍZ EN INTERVALO [-0.7, -0.3] CON MÉTODO DE REGLA FALSA:")
raiz1 = regla_falsa(-0.7, -0.3, 0.0001)
print(f"Raíz aproximada: {raiz1}\n")

# ---------------------------
# e) Raíz en intervalo [-0.2, 0.2]
# ---------------------------
print("RAÍZ EN INTERVALO [-0.2, 0.2] CON MÉTODO DE REGLA FALSA:")
raiz2 = regla_falsa(-0.2, 0.2, 0.0001)
print(f"Raíz aproximada: {raiz2}\n")

# ---------------------------
# f) Raíz en intervalo [0.5, 0.7]
# ---------------------------
print("RAÍZ EN INTERVALO [0.5, 0.7] CON MÉTODO DE REGLA FALSA:")
raiz3 = regla_falsa(0.5, 0.7, 0.0001)
print(f"Raíz aproximada: {raiz3}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import newton

# Definir las funciones
def f1(x):
    return np.sin(x) + x**2

def f2(x):
    return np.sin(x) - x**2

# a) Graficar y = sen(x) + x^2
x = np.linspace(-2, 2, 400)
y1 = f1(x)
plt.figure(figsize=(8,5))
plt.plot(x, y1, label='y = sin(x) + x^2')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.grid()
plt.show()

# b) Graficar y = sen(x) - x^2
x = np.linspace(-2, 2, 400)
y2 = f2(x)
plt.figure(figsize=(8,5))
plt.plot(x, y2, label='y = sin(x) - x^2', color='red')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.grid()
plt.show()

# c) Programa para encontrar las raíces de y = sen(x) - x^2
# d) y e) Encontrar raíces con Newton-Raphson
def df2(x):
    return np.cos(x) - 2*x  # Derivada de f2(x)

def find_root(x0, tol=0.0001):
    root = newton(f2, x0, df2, tol=tol)
    return root

# d) Para xn-1 = 1.3
root1 = find_root(1.3)
print(f'Raíz aproximada para x0 = 1.3: {root1}')

# e) Para xn-1 = -0.5
root2 = find_root(-0.5)
print(f'Raíz aproximada para x0 = -0.5: {root2}')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import newton

# Definir la función y su derivada
def f(x):
    return np.tan(x) - x

def df(x):
    return 1 / np.cos(x)**2 - 1  # Derivada de f(x)

# a) y b) Graficar la función y = tan(x) - x en el rango -2 <= x <= 2
x = np.linspace(-2, 2, 400)
y = f(x)
plt.figure(figsize=(8,5))
plt.plot(x, y, label='y = tan(x) - x')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.ylim(-10, 10)  # Para evitar valores extremos en las asíntotas
plt.legend()
plt.grid()
plt.show()

# c) Programa para encontrar la raíz usando Newton-Raphson
def find_root(x0, tol=0.0001):
    root = newton(f, x0, df, tol=tol)
    return root

# d) Encontrar la raíz con x0 = -1.2
root1 = find_root(-1.2)
print(f'Raíz aproximada para x0 = -1.2: {root1}')

# e) Encontrar la raíz con x0 = 1.2
root2 = find_root(1.2)
print(f'Raíz aproximada para x0 = 1.2: {root2}')
print("El valor obtenido no está cerca de la raíz porque la función tiene asíntotas, lo que puede causar divergencia en el método de Newton-Raphson.")

# f) Encontrar la raíz con x0 = 0
root3 = find_root(0)
print(f'Raíz aproximada para x0 = 0: {root3}')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import root_scalar

# Definir la función
def f(x):
    return 5 * np.sin(x) - x**3

# a) y b) Graficar la función en el intervalo -2 <= x <= 2
x = np.linspace(-2, 2, 400)
y = f(x)
plt.figure(figsize=(8,5))
plt.plot(x, y, label='y = 5sin(x) - x^3')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.grid()
plt.show()

# c) Programa para encontrar raíces usando el método de la secante
def secant_method(f, x0, x1, tol=0.0001):
    result = root_scalar(f, method='secant', x0=x0, x1=x1, xtol=tol)
    if result.converged:
        return result.root
    else:
        return None

# d) Encontrar la raíz en el intervalo [-2, -1.5]
root1 = secant_method(f, -2, -1.5)
print(f'Raíz aproximada en [-2, -1.5]: {root1}')

# e) Encontrar la raíz en el intervalo [-0.5, 0.5]
root2 = secant_method(f, -0.5, 0.5)
print(f'Raíz aproximada en [-0.5, 0.5]: {root2}')

# f) Encontrar la raíz en el intervalo [1.6, 1.9]
root3 = secant_method(f, 1.6, 1.9)
print(f'Raíz aproximada en [1.6, 1.9]: {root3}')

# g) Encontrar la raíz en el intervalo [1, 1.5]
root4 = secant_method(f, 1, 1.5)
print(f'Raíz aproximada en [1, 1.5]: {root4}')
print("La raíz encontrada puede no estar en el intervalo [1,1.5] debido a la naturaleza del método de la secante, que no siempre garantiza la convergencia dentro del intervalo dado.")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import root_scalar

# Definir la función
def f(x):
    return 5*x**2 + 6*x - x**3

# a) y b) Graficar la función en el intervalo 0 <= x <= 2.5
x = np.linspace(0, 2.5, 400)
y = f(x)
plt.figure(figsize=(8,5))
plt.plot(x, y, label='y = 5x^2 + 6x - x^3')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.grid()
plt.show()

# c) Programa para encontrar raíces usando el método de la secante
def secant_method(f, x0, x1, tol=0.0001):
    result = root_scalar(f, method='secant', x0=x0, x1=x1, xtol=tol)
    if result.converged:
        return result.root
    else:
        return None

# d) Encontrar la raíz en el intervalo [1.8, 2]
root1 = secant_method(f, 1.8, 2)
print(f'Raíz aproximada en [1.8, 2]: {root1}')

# e) Explicación de la no convergencia en el intervalo [0.25, 0.5]
root2 = secant_method(f, 0.25, 0.5)
if root2 is None:
    print('El método de la secante no converge en el intervalo [0.25, 0.5].')
else:
    print(f'Raíz aproximada en [0.25, 0.5]: {root2}')
```

```python
# Algoritmo de Brent para encontrar la raíz de la función f(x) = sqrt(5x^2 + 6x) - x^3
import math

def f(x):
    valor = 5 * x**2 + 6*x
    if valor < 0:
        return float('nan')  # Evita valores complejos
    return math.sqrt(valor) - x**3

def brent_method(a, b, tol=1e-4, max_iter=100):
    if f(a) * f(b) >= 0 or math.isnan(f(a)) or math.isnan(f(b)):
        raise ValueError("Los valores iniciales no encierran una raíz o generan valores inválidos")

    fa, fb = f(a), f(b)
    c, fc = a, fa
    d, e = 0, 0

    for _ in range(max_iter):
        if fb == 0 or abs(a - b) < tol:
            return b

        if fa != fc and fb != fc:
            # Interpolación inversa cuadrática
            s = a * fb * fc / ((fa - fb) * (fa - fc)) + \
                b * fa * fc / ((fb - fa) * (fb - fc)) + \
                c * fa * fb / ((fc - fa) * (fc - fb))
        else:
            # Método de la secante
            s = b - fb * (b - a) / (fb - fa)

        cond1 = (3 * a + b) / 4
        cond2 = (s < cond1 or s > b)
        cond3 = (e and abs(s - b) >= abs(b - c) / 2)
        cond4 = (not e and abs(s - b) >= abs(c - d) / 2)
        cond5 = (e and abs(b - c) < tol)
        cond6 = (not e and abs(c - d) < tol)

        if cond2 or cond3 or cond4 or cond5 or cond6 or math.isnan(f(s)):
            s = (a + b) / 2
            e, d = b - a, s - b

        fs = f(s)
        d, c, fc = c - b, b, fb

        if fa * fs < 0:
            b, fb = s, fs
        else:
            a, fa = s, fs

        if abs(fa) < abs(fb):
            a, b = b, a
            fa, fb = fb, fa

    return b  # Aproximación final

# Prueba del algoritmo
try:
    root = brent_method(1.8, 2, tol=0.0001)
    print(f"Raíz aproximada: {root}")
except ValueError as e:
    print(f"Error: {e}")
```
